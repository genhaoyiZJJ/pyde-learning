#### 展示缺失值

```python
def missing_value_table(df):
    #计算所有的缺失值
    mis_val = df.isnull().sum() # 每列的缺失值总数
    # 缺失值百分比
    mis_val_percent = 100*df.isnull().sum()/len(df)
    #合并
    mis_val_table = pd.concat([mis_val,mis_val_percent],axis=1)
    mis_val_rename = mis_val_table.rename(columns = {0:'Missing valyes',1:'% of total values'})
    #将没有缺失值的列删除，只显示有缺失的列
    mis_val_rename = mis_val_rename[mis_val_rename.iloc[:,1]!=0].sort_values('% of total values',ascending=False)
    return mis_val_rename
```

### 特征工程

```python
poly_features = app_train[['TARGET','EXT_SOURCE_1','EXT_SOURCE_2','EXT_SOURCE_3','DAYS_BIRTH']]
from sklearn.preprocessing import PolynomialFeatures
from sklearn.impute import SimpleImputer
```

```python
#用特征的中位数填充缺失值
imputer = SimpleImputer(strategy='median')
poly_target = poly_features['TARGET']
poly_features.drop(columns=['TARGET'],inplace =True)
poly_features = imputer.fit_transform(poly_features)
```

```python
# 创建一个 PolynomialFeatures 对象，用于将特征转换为多项式特征。degree=3 表示我们希望将特征转换为三次多项式特征
poly_transformer = PolynomialFeatures(degree=3)
poly_transformer.fit(poly_features)
# 将 poly_features 数据中的特征转换为多项式特征
poly_features = poly_transformer.transform(poly_features)
```

```python
poly_features.shape
```

```python
poly_transformer.get_feature_names(input_features=['TARGET','EXT_SOURCE_1','EXT_SOURCE_2','EXT_SOURCE_3','DAYS_BIRTH'])[:20]
```

将当前得到的部分特征跟总体组合在一起

```python
poly_features = pd.DataFrame(poly_features,columns=poly_transformer.get_feature_names(input_features=['TARGET','EXT_SOURCE_1','EXT_SOURCE_2','EXT_SOURCE_3','DAYS_BIRTH']))
poly_features.head()
```

```python
poly_features['SK_ID_CURR'] = app_train['SK_ID_CURR']
app_train_poly = app_train.merge(poly_features,on='SK_ID_CURR',how='left')
app_train_poly.head()
```

根据实际情况来创建新的特征

```python
app_train_domain = app_train.copy()

app_train_domain['CREDIT_INCOME_PERCENT'] = app_train_domain['AMT_CREDIT'] / app_train_domain['AMT_INCOME_TOTAL']
app_train_domain['ANNUITY_INCOME_PERCENT'] = app_train_domain['AMT_ANNUITY'] / app_train_domain['AMT_INCOME_TOTAL']
app_train_domain['CREDIT_TERM'] = app_train_domain['AMT_ANNUITY'] / app_train_domain['AMT_CREDIT']
app_train_domain['DAYS_EMPLOYED_PERCENT'] = app_train_domain['DAYS_EMPLOYED'] / app_train_domain['DAYS_BIRTH']
```

```python
plt.figure(figsize = (12, 20))
for i, feature in enumerate(['CREDIT_INCOME_PERCENT', 'ANNUITY_INCOME_PERCENT', 'CREDIT_TERM', 'DAYS_EMPLOYED_PERCENT']):
    
    plt.subplot(4, 1, i + 1)
    sns.kdeplot(app_train_domain.loc[app_train_domain['TARGET'] == 0, feature], label = 'target == 0')
    sns.kdeplot(app_train_domain.loc[app_train_domain['TARGET'] == 1, feature], label = 'target == 1')
    
    plt.title('Distribution of %s by Target Value' % feature)
    plt.xlabel('%s' % feature); plt.ylabel('Density');
    
plt.tight_layout(h_pad = 2.5)
plt.show()
```

### 数据预处理

```python
from sklearn.preprocessing import StandardScaler,MinMaxScaler
from sklearn.model_selection import train_test_split

# 分为目标变量（label）和特征（train）
label = app_train['TARGET']
train = app_train.drop(columns = ['TARGET'])
# 将数据集分为训练集和测试集
train,test,y_train,y_test= train_test_split(train,label,test_size = 0.2,random_state = 0)
features = list(train.columns)

imputer = SimpleImputer(strategy='median')
std = StandardScaler()
#填充
imputer.fit(train)
train = imputer.transform(train)
test = imputer.transform(test)
#标准化
std.fit(train)
train = std.transform(train)
test = std.transform(test)
```

基础模型：逻辑回归

```python
from sklearn.linear_model import LogisticRegression
log_reg = LogisticRegression(C=0.0001)
log_reg.fit(train,y_train)
```
